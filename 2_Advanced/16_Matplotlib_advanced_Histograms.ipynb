```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title_short']=='Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US['salary_year_avg'].plot(kind='hist', bins=30, edgecolor='black')
plt.xlim(0, 250000)

ax=plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
```

# Practice Problems

```python
df_DS_US = df[(df['job_title_short']== 'Data Scientist') & (df['job_country']== 'United States')].copy()
```

```python
df_DS_US['salary_hour_avg'].plot(kind='hist', edgecolor='black', bins=30)
plt.xlim(0, 150)

ax=plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x)}'))
plt.title('Hourly Salary for Data Scientists in USA')
plt.xlabel('Hourly Pay')
plt.ylabel('Number of Job Postings')
```

```python
df['job_skills_cleaned'] = df['job_skills'].apply(lambda x: x if isinstance(x, list) else [])
```

```python
df['job_skills_count'] = df['job_skills_cleaned'].apply(lambda x: len(x))
```

```python
df['job_skills_count'].plot(kind='hist', edgecolor='black', bins=50)
plt.xlim(0, 20)
plt.title('Skills Per Job Posting')
plt.xlabel('Skills Posted')
plt.ylabel('Job Postings')
```
