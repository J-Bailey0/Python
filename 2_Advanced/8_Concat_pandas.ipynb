```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
```

```python
months = df['job_posted_month'].unique()
{month: month for month in months}
```

```python
# This is looping through all the different months and adding that info to a dictonary
dict_months = {month: df[df['job_posted_month'] == month] for month in months}
```

```python
countries = df['job_country'].unique()
```

```python
dict_country = {country: df[df['job_country'] == country] for country in countries}
```

```python
dict_country['Mexico']
```

```python
dict_months['Jan']
```

```python
# Boss needs Q1 Data
df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)
```

```python
df_north_america = pd.concat([dict_country['United States'], dict_country['Mexico'], dict_country['Canada']], ignore_index=True)
```

```python
df_q1['job_posted_month'].value_counts().plot(kind='line')
```

```python
df_north_america.drop_duplicates()
```

```python
df_north_america['job_country'].value_counts().plot(kind='barh')
```

```python
df_north_america.dropna(subset=['salary_year_avg'], inplace=True)
```

```python
df_north_america_pivot = df_north_america.pivot_table(index='job_country', values='salary_year_avg', aggfunc='median')
```

```python
df_north_america_pivot.reset_index(inplace=True)
```

```python
df_north_america_pivot.set_index('job_country', inplace=True)
```

```python
colors = ['red' if country == 'Canada' else 'orange' for country in df_north_america_pivot.index]
df_north_america_pivot.plot(kind='bar', color=colors)
plt.ylim(0, 200000)
```
