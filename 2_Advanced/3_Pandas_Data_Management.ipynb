```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
#Cleaning Data
df['job_posted_date']  = pd.to_datetime(df['job_posted_date'])
```

```python
df_altered = df.copy()

median_salary_year = df_altered['salary_year_avg'].median()
median_salary_hour = df_altered['salary_hour_avg'].median()

df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary_year)
df_altered['salary_hour_avg'] = df_altered['salary_hour_avg'].fillna(median_salary_hour)
```
